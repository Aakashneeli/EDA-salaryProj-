```python
#building a word cloud for our job descriptions, (word cloud basically is a box with keywords that are most occuring in the dataset)
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import nltk
```

```python
words = ''.join(df['Job Description'])

#punctuation remover
def remove_punc(text):
    # remove punctuation marks using string.punctuation
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = remove_punc(words)

text = "".join([ele for ele in words_filtered])[:10000]

wc =  WordCloud(background_color = 'white', random_state = 1, stopwords = STOPWORDS ,max_words = 2000, width = 800, height= 1500)
wc.generate(text)

plt.figure(figsize = [10, 10])
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
